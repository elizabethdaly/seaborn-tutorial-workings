## Visualizing statistical relationships

The tips dataset

```python
# Import all the packages required.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
# The tips dataset comes with Seaborn. Import it to a dataframe
tips = sns.load_dataset("tips")
#tips.sample(7)
#tips.info()
sns.relplot(x="total_bill", y="tip", data=tips);
```

Add a third variable to the plot in form of smoker/non-smoker, each with different coloured markers. Could do it by day of week.

```python
# hue keyword adds extra information to the plot.
sns.relplot(x="total_bill", y="tip", hue="day", data=tips);
```

Modify the third variable smoker/non-smoker to have different colour and different markers.

```python
# Use a different marker style.
sns.relplot(x="total_bill", y="tip", hue="day", style="smoker", data=tips);
```

Can even add a fourth variable with style and colour set independently.

```python
sns.relplot(x="total_bill", y="tip", hue="day", style="time", data=tips);
```

If hue semantic is numeric (eg size), pallatte changes to a sequential one.

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

Can also change the size of each point. size="smoker", "sex" for eg also works.

```python
sns.relplot(x="total_bill", y="tip", size="sex", data=tips);
```

## Building structured multiplot grids

```python
g = sns.FacetGrid(tips, col="time")
g.map(plt.hist, "tip");
```

```python
g = sns.FacetGrid(tips, col="sex", hue="smoker")
g.map(plt.scatter, "total_bill", "tip", alpha=.7)
g.add_legend();
```

```python
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill");
```

```python
 ordered_days = tips.day.value_counts().index
g = sns.FacetGrid(tips, row="day", row_order=ordered_days,
                  height=1.7, aspect=4,)
g.map(sns.distplot, "total_bill", hist=False, rug=True);
```

```python
pal = dict(Yes="seagreen", No="gray")
g = sns.FacetGrid(tips, hue="smoker", palette=pal, height=5)
g.map(plt.scatter, "total_bill", "tip", s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.add_legend();
```

```python
g = sns.FacetGrid(tips, hue="sex", palette="Set1", height=5, hue_kws={"marker": ["^", "v"]})
g.map(plt.scatter, "total_bill", "tip", s=100, linewidth=.5, edgecolor="white")
g.add_legend();
```
